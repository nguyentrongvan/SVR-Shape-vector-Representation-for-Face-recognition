```python
import cv2
import numpy as np

import os
from tqdm import tqdm
import matplotlib.pyplot as plt
```

```python
path='olivetti_segment/'
list_img=os.listdir(path)
len(list_img)
```

```python
exp_img=cv2.imread('olivetti_segment/1_13.jpg',0)
h,w=exp_img.shape[0],exp_img.shape[1]

plt.imshow(exp_img)
plt.title('Height: {} - Width: {}'.format(h,w))
plt.axis('off')
```

```python
def getVector(list_file):
    data=[]
    label=[]
    
    for file in tqdm(list_file):
        img=cv2.imread(path+file,0)
        vector=np.reshape(img,(256*256))
        data.append(vector)
        label.append(file.split('_')[0])
        
    np.save('vectors_oli.npy',data)
    np.save('labels_oli.npy',label)
```

```python
getVector(list_img)
```
